```python
# First let's import the packages we will use in this project

import pandas as pd
import numpy as np
import seaborn as sns
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



```

```python
# Now we need to read in the data
df = pd.read_csv('/Users/saad/Desktop/movies.csv')
```

```python
df.head()
```

```python
# See if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
newDf = df.dropna(axis=0)

for col in newDf.columns:
    pct_missing = np.mean(newDf[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data Types for our columns

newDf.dtypes
```

```python
# Change data type of columns

newDf['budget'] = newDf['budget'].astype('int64')
newDf['gross'] = newDf['gross'].astype('int64')
```

```python
newDf
```

```python
newDf.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
 # Drop any duplicates
    
newDf.drop_duplicates()
```

```python
# Scatter plot with budget vs gross

plt.scatter(x=newDf['budget'], y=newDf['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget For Film')

plt.show()
```

```python
newDf.head()
```

```python
# Plot budget VS gross using seaborn

sns.regplot(x="gross", y="budget", data=newDf, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# Correlation Matrix between all numeric columns

newDf.corr(method ='pearson')
```

```python
newDf.corr(method ='kendall')
```

```python
newDf.corr(method ='spearman')
```

```python
correlation_matrix = newDf.corr(method ='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
df_numerized = newDf


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
correlation_matrix = df_numerized.corr(method ='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
df_numerized.corr(method='pearson')
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
